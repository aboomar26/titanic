```python
from sklearn.base import BaseEstimator,TransformerMixin
```

```python
from sklearn.impute import SimpleImputer
class Ageimputer (BaseEstimator,TransformerMixin):
    def fit (self,X,y=None):
        return self
    
    def transform(self,X):
        imputer=SimpleImputer(strategy="mean")
        X['Age']=imputer.fit_transform(X[['Age']])
        
        return X
```

```python
from sklearn.preprocessing import OneHotEncoder

class featureencode (BaseEstimator,TransformerMixin):
    def fit(self ,X,y=None):
        return self
    
    def transform(self,X):
        encoder=OneHotEncoder()
        matrix=encoder.fit_transform(X[['Embarked']]).toarray()
        
        column=['C','S','Q','N']
        
        for i in range (len(matrix.T)):
            X[column[i]]=matrix.T[i]
            
            
        matrix=encoder.fit_transform(X[["Sex"]]).toarray()
        
        columns=["Male","Female"]
        
        for i in range(len(matrix.T)):
            X[columns[i]]=matrix.T[i]
            
        return X
```

```python
class dropfeature (BaseEstimator,TransformerMixin):

    def fit(self,X,y=None):
        return self
    
    def transform(self,X):
        return X.drop(["Sex","Name","Ticket","N","Embarked","Cabin"],axis=1)
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipeline=Pipeline([('featureencode',featureencode()),
                  ("Ageimputer",Ageimputer()),
                  ('dropfeature',dropfeature()),
                  ])
```
